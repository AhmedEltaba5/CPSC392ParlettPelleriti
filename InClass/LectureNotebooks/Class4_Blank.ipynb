```python
iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print(iris.head())
```

# GG Basic Steps

```python
(ggplot(iris, aes("sepal_width", "sepal_length", color = "species")))
```

## 2. Tell GG HOW you want the data to show up

```python
(ggplot(iris, aes("sepal_width", "sepal_length", color = "species"))
+ geom_point())
```

## 1. Tell GG WHAT data you want to use

## 1. Tell GG WHAT data you want to use

```python
(ggplot(iris, aes("species"))
)
```

## 2. Tell GG HOW you want the data to show up

```python
(ggplot(iris, aes("species"))
 + geom_bar())
```

## 1. Tell GG WHAT data you want to use

## 2. Tell GG HOW you want the data to show up

```python
(ggplot(iris, aes(x = 'species', y = 'sepal_length')) 
+ geom_boxplot())
```

## 3. Extras

```python
(ggplot(iris, aes(x = 'species', y = 'sepal_length', fill = "species")) 
+ geom_boxplot())
```

```python
(ggplot(iris, aes(x = 'species', y = 'sepal_length', fill = "species")) 
+ geom_boxplot() 
+ theme_minimal())
```

```python
(ggplot(iris, aes(x = 'species', y = 'sepal_length', fill = "species")) 
+ geom_boxplot() 
+ theme_minimal() 
+ geom_point())
```

# Prepping Data for Barcharts


```python
bar_iris = iris.groupby("species").agg({"sepal_width": ["mean", "std", "min"]})
bar_iris.columns = ["height_cm", "std", "min"]
bar_iris["species"] = bar_iris.index
bar_iris
```

```python
(ggplot(bar_iris, aes(x = "species", y = "sepal_width", fill = "species"))
+ geom_bar(stat = "identity"))
```
