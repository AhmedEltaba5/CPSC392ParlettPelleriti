```python
(ggplot(mtcars, aes('wt', 'mpg'))
 + geom_point())
```

## 1. Tell GG WHAT data you want to use

## 2. Tell GG HOW you want the data to show up

```python
(ggplot(penguin, aes(x = "species",
                        y = "bill_length_mm")) + geom_boxplot())
```

## 3. Extras

```python
(ggplot(penguin, aes(x = "species",
                        y = "bill_length_mm",
                    fill = "species")) + geom_boxplot())
```

```python
(ggplot(penguin, aes(x = "species",
                        y = "bill_length_mm",
                    fill = "species")) + geom_boxplot() + theme_minimal())
```

```python
(ggplot(penguin, aes(x = "species",
                        y = "bill_length_mm", fill = "species")) + geom_boxplot() +
 theme_minimal() + geom_point())
```

```python
(ggplot(penguin, aes(x = "species",
                        y = "bill_length_mm")) + geom_boxplot(aes(fill = "species")) +
 theme_minimal() + geom_point())
```

# Prepping Data for Barcharts


```python
(ggplot(penguin, aes( x = "species", y = "body_mass_g", fill = "species")) +
 stat_summary(fun_data = "mean_sdl", geom = "bar"))
```

```python
(ggplot(penguin, aes( x = "species", y = "body_mass_g", fill = "species")) +
 stat_summary(fun_data = "mean_sdl", geom = "bar")+
 labs(x = "Species", y = "Body Mass (g)") + 
 ggtitle("Penguin Body Mass by Species") +
 theme_minimal() +
 theme(panel_grid_major_x = element_blank(),
     panel_grid_minor_x = element_blank(),
      panel_grid_minor_y = element_blank(),
      legend_position = "none"))
```
