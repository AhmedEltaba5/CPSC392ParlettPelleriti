```python
t = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/telecom_churn.csv")
print(t.shape)
print(t.columns)
t.head()

```

```python
#2
feat = ["number vmail messages", "total day minutes", "total day calls", 
       "total eve minutes", "total eve calls", "total night minutes",
       "total night calls",
       "customer service calls"]

X = t[feat]
y = t["churn"]

kf = KFold(n_splits = 4)

knn = KNeighborsClassifier()

acc = []

for train, test in kf.split(X,y):
    X_train = X.iloc[train]
    X_test = X.iloc[test]
    y_train = y[train]
    y_test = y[test]
    
    z = StandardScaler()
    X_train[feat] = z.fit_transform(X_train)
    X_test[feat] = z.transform(X_test)
    
    knn.fit(X_train,y_train)
    acc.append(knn.score(X_test,y_test))
    
    plot_confusion_matrix(knn, X_test,y_test)
print(np.mean(acc))
```
