# Model

```python
predictors = ["tiredness", "timesincemeal", "age"]
X = d[predictors]
y = d["reactionTime"]

print(d.shape)

kf = KFold(n_splits = 10)
kf.split(X)

mod = LinearRegression()
r2 = []
for train_i, test_i in kf.split(X):
    X_train = X.iloc[train_i]
    X_test = X.iloc[test_i]
    y_train = y[train_i]
    y_test = y[test_i]
    
    z = StandardScaler()
    z.fit(X_train)
    Xz_train = z.transform(X_train)
    Xz_test = z.transform(X_test)
    
    mod.fit(Xz_train,y_train)
    
    r2.append(mod.score(Xz_test,y_test))
    
print(r2)
print(np.mean(r2))
```
