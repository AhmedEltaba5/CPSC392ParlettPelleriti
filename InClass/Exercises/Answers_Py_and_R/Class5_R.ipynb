```r
library(ggplot2)
```

```r
df <- read.csv(url("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/heart.csv"))
```

```r
head(df)
```

```r
# BAD GRAPH
ggplot(df, aes(fbs, restecg)) + geom_point() + labs(x = "", y = "")
```

This graph is bad because it does not tell the viewer what the graph is about. It also represents categorical variables as continuous, using a scatter plot. This is not a good type of graph for this data. When looking at the relationship between 2 categories, we often want to know the frequency of combinations of categories. This graph just shows you that 5 combinations exist (the 5 data points).

```r
# Good Graph

ggplot(df, aes(factor(fbs), fill = factor(restecg))) + geom_bar() +
labs(x = "Fasting Blood Sugar > 120 mg/dl (0-False, 1-True)", title = "Restig ECG and Fasting Blood Sugar in Potential Heart Disease Cases") +
scale_fill_discrete(name = "Resting ECG (0-Normal)") +
theme_classic()
```

```r
This graph is good because it shows us the proportion of subjects in each combination of categories. It also has clear axis labels and a good title. We also got rid of extra details using theme_classic() so that people can focus more on the actual graph. We used factor() to change the representation of the variables from continuous to categorical.

For example, notice that for Resting ECG, group 2 ("showing probable or definite left ventricular hypertrophy by Estes' criteria"), only exists when people do not have a Fasting Blood Sugar > 12 mg/dl. That indicates that we should look more into the relationship between those two things.
```
