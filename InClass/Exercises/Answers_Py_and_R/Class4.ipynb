```python
import warnings
warnings.filterwarnings('ignore')


import pandas as pd
from plotnine import *
```

# 2. Load the following dataset . For more info check out [this link.](https://www.kaggle.com/stefanoleone992/fifa-20-complete-player-dataset/version/1#)

# 4. How old is the average player in FIFA20? Plot a histogram of their ages.

```python
Fifa["age"].mean()
```

```python
(ggplot(Fifa, aes("age")) + geom_histogram())
```

# 5. Do right footed players weigh more than right footed players? Use ggplot/plotnine to make a graph to answer this question.

```python
(ggplot(Fifa, aes(x = "preferred_foot", y = "weight_kg")) + geom_boxplot(aes(fill = "preferred_foot")))
```

# 6. Is there a difference in height between the clubs? Use ggplot/plotnine to make a graph to answer this question.

```python
clubBool = (Fifa["club"] == "Chelsea") | (Fifa["club"] == "Manchester United")  
Chelsea_Manchester = Fifa.loc[clubBool]
Chelsea_Manchester.head()
```

```python
(ggplot(Chelsea_Manchester, aes(x = "club", y = "height_cm")) + geom_boxplot(aes(fill = "club")))
```

```python
clubMeans = Chelsea_Manchester.groupby("club").mean()[["height_cm"]]
clubMeans
```

```python
clubMeans["club"] = list(clubMeans.index)
clubMeans
```

```python
#another way
clubMeans2 = Chelsea_Manchester.groupby("club").agg({'height_cm': ["mean"]})
clubMeans2
```

```python
(ggplot(clubMeans, aes(x = "club", y = "height_cm")) + geom_bar(aes(fill = "club"), stat = "identity"))
```
