## 1. Interpreting Linear Regression

Use the *kc_house_data.csv* (more info about the variables here: https://www.kaggle.com/harlfoxem/housesalesprediction) to build a linear regression model that looks at the price of houses in King County.

* Which variables are you using as predictors? Why?
* Standardize
* What kind of cross validiation do you want to use here?

```python
kc = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/kc_house_data.csv")
print(kc.shape)
kc.head()
```

```python
feat = ["bedrooms", "bathrooms", "sqft_living", "view"]

X = kc[feat]
y = kc["price"]

kf = KFold(5)
lr = LinearRegression()
r2 = []

for train, test in kf.split(X):
    X_train = X.iloc[train]
    X_test = X.iloc[test]
    y_train = y[train]
    y_test = y[test]
    
    z = StandardScaler()

    X_train[["sqft_living"]] = z.fit_transform(X_train[["sqft_living"]])
    X_test[["sqft_living"]] = z.transform(X_test[["sqft_living"]])
    
    lr.fit(X_train,y_train)
    r2.append(lr.score(X_test, y_test))
    
print(np.mean(r2))
    
    
```

## 3. Interpreting Logistic Regression
Using the *BreastCancer.csv* data, build a model that predicts whether the breast cancer biopsy is benign (B) or malignant (M). (more info about the data here: https://www.kaggle.com/uciml/breast-cancer-wisconsin-data)

* Which variables are you using as predictors? Why?
* Standardize
* What kind of cross validiation do you want to use here?

```python
bc = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/BreastCancer.csv")
print(bc.shape)
bc.head()
```

## 4. Evaluating Logistic Regression Model Performance
How well did your model perform? How do you know?

```python
feat = ["radius_mean", "texture_mean", "perimeter_mean",
        "area_mean", "smoothness_mean", "compactness_mean", "concavity_mean"]

X = bc[feat]
y = bc["diagnosis"]

b = LabelBinarizer()

y = b.fit_transform(y)

loo = LeaveOneOut()
acc = []

logr = LogisticRegression()

for train, test in loo.split(X):
    X_train = X.iloc[train]
    X_test = X.iloc[test]
    y_train = y[train]
    y_test = y[test]
    
    z = StandardScaler()
    z.fit(X_train)
    
    Xz_train = z.transform(X_train)
    Xz_test = z.transform(X_test)
    
    logr.fit(Xz_train,y_train)
    
    acc.append(logr.score(Xz_test,y_test))
#     plot_confusion_matrix(logr, Xz_test, y_test)
  
print(np.mean(acc))
```
