# 6. Is there a difference in height between the clubs? Use ggplot/plotnine to make a graph to answer this question.


```r
clubBool <- Fifa$club == "Chelsea" | Fifa$club == "Manchester United"  
Chelsea_Manchester <- Fifa[clubBool,]
head(Chelsea_Manchester)
```

```r
ggplot(Chelsea_Manchester, aes(x = club, y = height_cm)) + geom_boxplot(aes(fill = club))
```

```r
clubMeans <- ddply(Chelsea_Manchester,~club,summarise,height_cm=mean(height_cm)) #uses plyr
clubMeans
```

```r
ggplot(clubMeans, aes(x = club, y = height_cm)) + geom_bar(aes(fill = club), stat = "identity")
```
