```python
### YOUR CODE HERE ###
fifa = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/players_15.csv")

# get month of date of birth
fifa["monthBorn"] = fifa["dob"].str.extract(r'-([0-9][0-9])-')

fifa.head()
```

```python
#3
ggplot(fifa, aes("age")) + geom_histogram(fill = "lightblue", color = "black") + theme_minimal()

#or

ggplot(fifa, aes("age")) + geom_density(fill = "lightblue", color = "black") + theme_minimal()
```

```python
#4
ggplot(fifa, aes(x = "preferred_foot", y = "weight_kg", fill = "preferred_foot")) + geom_boxplot() + theme_minimal()
```

```python
#5

CorManU = (fifa["club"] == "Chelsea") | (fifa["club"] == "Manchester United")

CandManU = fifa.loc[CorManU]

ggplot(CandManU, aes(x = "club", y = "height_cm")) + geom_boxplot()
```

```python
#6

ggplot(fifa, aes(x = "monthBorn", y = "height_cm", fill = "height_cm")) + geom_boxplot()
```

```python
#7

ggplot(fifa, aes(x = "body_type", y = "weight_kg", fill = "body_type")) + stat_summary(fun_data = "mean_sdl", geom = "bar")
```

```python
#8

height_sq = (fifa["height_cm"] * 100)**2

fifa["bmi"] = fifa["weight_kg"]/height_sq

(ggplot(fifa, aes(x = "bmi")) + geom_histogram(fill = "white", color = "black") +
geom_vline(xintercept = fifa["bmi"].mean() , linetype = "dashed", size = 3) +
theme_minimal())
```

```python
#9
(ggplot(fifa, aes(x = "height_cm", y = "weight_kg")) + geom_point(alpha = 0.2) +
 geom_smooth(method = "lm", color = "blue") +  facet_wrap("~body_type") + theme_minimal())
```

```python
#10

ggplot(fifa, aes(x = "team_jersey_number", y = "age")) + geom_point()
```

```python
#11
# Let's use your CPSC230 skills, create a column in fifa called name_len that
# counts the number of characters in each player's long_name (spaces shouldn't count).
# Then use ggplot/plotnine to create a histogram of name_len and add +theme_minimal()
# to your graph. What is the typical range of name length?

fifa["name_len"] = [len("".join(name.split())) for name in fifa["long_name"]]

ggplot(fifa, aes(x = "name_len")) + geom_histogram() + theme_minimal()
```

The typical range of name_len is between 5 and about 25 characters
